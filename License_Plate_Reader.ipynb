```python
import cv2
from easyocr import Reader
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('default')

import numpy as np
from imutils import grab_contours
import os
import pytesseract as PT
PT.pytesseract.tesseract_cmd = os.path.join(os.getenv('LOCALAPPDATA'), 'Tesseract-OCR\\tesseract.exe')

import re
import warnings
warnings.filterwarnings("ignore")
```

```python
img = cv2.imread(os.path.join(os.path.abspath(''), 'images\\car3.jpg'))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
alpha = 1.2 # Contrast control (1.0-3.0)
beta = 0 # Brightness control (0-100)

adjusted = cv2.convertScaleAbs(img, alpha = alpha, beta = beta)
gray = cv2.cvtColor(adjusted, cv2.COLOR_BGR2GRAY) #convert to grey scale

plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
bfilter = cv2.bilateralFilter(gray, 10, 10, 10) # reduce noise
edged = cv2.Canny(bfilter, 100, 300) # edge detection

keypoints = cv2.findContours(edged, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
contours = grab_contours(keypoints)
contours = sorted(contours, key = cv2.contourArea, reverse = True)[:10]

location = None
for contour in contours:
    approx = cv2.approxPolyDP(contour, 10, True)
    if len(approx) == 4:
        location = approx
        break
        
mask = np.zeros(gray.shape, np.uint8)
new_image = cv2.drawContours(mask, [location], 0, 255, -1)
new_image = cv2.bitwise_and(img, img, mask = mask)

plt.imshow(cv2.cvtColor(new_image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
(x, y) = np.where(mask == 255)
(x1, y1) = (np.min(x), np.min(y))
(x2, y2) = (np.max(x), np.max(y))
cropped_image = gray[x1:x2+1, y1:y2+1]
plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
alpha = 2 # Contrast control (1.0-3.0)
beta = 25 # Brightness control (0-100)

adjusted_crop = cv2.convertScaleAbs(cropped_image, alpha = alpha, beta = beta)
plt.imshow(cv2.cvtColor(adjusted_crop, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Gaussian blur
blur = cv2.GaussianBlur(adjusted_crop, (3, 3), 1)
plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
reader = Reader(['en'])
result = reader.readtext(blur)

print(f"EasyOCR result: {result[0][1]}")
print(f"EasyOCR result confidence: {result[0][2]:,.2%}")
```
