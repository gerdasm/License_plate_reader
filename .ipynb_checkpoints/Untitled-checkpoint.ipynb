```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import imutils
import easyocr
```

```python
img = cv2.imread('3.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB));
```

```python
bfilter = cv2.bilateralFilter(gray, 15, 15, 15) # reduce noice
edged = cv2.Canny(bfilter, 100, 100) # edge detection
plt.imshow(cv2.cvtColor(edged, cv2.COLOR_BGR2RGB));
```

```python
keypoints = cv2.findContours(edged.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
contours = imutils.grab_contours(keypoints)
contours = sorted(contours, key = cv2.contourArea, reverse = True)[:10]
```

```python
location = None
for contour in contours:
    approx = cv2.approxPolyDP(contour, 10, True)
    if len(approx) == 4:
        location = approx
        break
location
```

```python
mask = np.zeros(gray.shape, np.uint8)
new_image = cv2.drawContours(mask, [location], 0, 255, -1)
new_image = cv2.bitwise_and(img, img, mask = mask)
plt.imshow(cv2.cvtColor(new_image, cv2.COLOR_BGR2RGB));
```

```python
(x, y) = np.where(mask == 255)
(x1, y1) = (np.min(x), np.min(y))
(x2, y2) = (np.max(x), np.max(y))
cropped_image = gray[x1:x2+1, y1:y2+1]
plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB));
```

```python
reader = easyocr.Reader(['en'])
result = reader.readtext(cropped_image)
result
```

```python
# text = result[0][-2]
text = result[1][-2] + result[0][-2]

font = cv2.FONT_HERSHEY_SIMPLEX
red = cv2.putText(img, text = text, org = (approx[0][0][0], approx[1][0][1]-100),
                  fontFace = font, fontScale = 2, color = (0, 255, 0), thickness = 3, lineType = cv2.LINE_AA)
res = cv2.rectangle(img, tuple(approx[0][0]), tuple(approx[2][0]), (0,255,0), 3)
plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB));
```
