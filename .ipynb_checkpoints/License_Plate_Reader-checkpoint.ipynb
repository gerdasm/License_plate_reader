```python
import cv2
from easyocr import Reader
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('default')

import numpy as np
from imutils import grab_contours
import os
import pytesseract as PT
PT.pytesseract.tesseract_cmd = os.path.join(os.getenv('LOCALAPPDATA'), 'Tesseract-OCR\\tesseract.exe')

import re
import warnings
warnings.filterwarnings("ignore")
```

```python
img = cv2.imread(os.path.join(os.path.abspath(''), 'images\\car3.jpg'))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
alpha = 1 # Contrast control (1.0-3.0)
beta = 0 # Brightness control (0-100)

adjusted = cv2.convertScaleAbs(img, alpha = alpha, beta = beta)
gray = cv2.cvtColor(adjusted, cv2.COLOR_BGR2GRAY)

plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
bfilter = cv2.bilateralFilter(gray, 10, 10, 10) # reduce noise
edged = cv2.Canny(bfilter, 250, 500) # edge detection
plt.imshow(cv2.cvtColor(edged, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
keypoints = cv2.findContours(edged, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
contours = grab_contours(keypoints)
contours = sorted(contours, key = cv2.contourArea, reverse = True)[:10]

location = None
for contour in contours:
    approx = cv2.approxPolyDP(contour, 10, True)
    if len(approx) == 4:
        location = approx
        break
        
mask = np.zeros(gray.shape, np.uint8)
new_image = cv2.drawContours(mask, [location], 0, 255, -1)
new_image = cv2.bitwise_and(img, img, mask = mask)

(x, y) = np.where(mask == 255)
(x1, y1) = (np.min(x), np.min(y))
(x2, y2) = (np.max(x), np.max(y))
cropped_image = gray[x1:x2+1, y1:y2+1]
plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Gaussian blur
blur = cv2.GaussianBlur(cropped_image, (5, 5), 1)
ret3, th3 = cv2.threshold(blur,0,255,cv2.THRESH_BINARY + cv2.THRESH_OTSU)
plt.imshow(cv2.cvtColor(th3, cv2.COLOR_BGR2RGB))
plt.show()

# manual thresholding
th2 = 100 # this threshold might vary!
blur[blur >= th2] = 255
blur[blur < th2]  = 0

plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
reader = Reader(['en'], gpu = False)
result = reader.readtext(blur, paragraph = "False")

easy_ocr_text = re.sub('[^A-Za-z0-9]+', '', result[0][1]).upper()
print('EasyOCR result:', easy_ocr_text)
print("PyTesseract result:", re.sub('[^A-Za-z0-9]+', '', PT.image_to_string(cropped_image, config="--psm 6")).upper())
```

```python
text = easy_ocr_text ## EasyOCR result

font = cv2.FONT_HERSHEY_SIMPLEX

rec = cv2.rectangle(img, tuple(approx[0][0]), tuple(approx[2][0]), (0,255,0), 3)
res = cv2.putText(img = rec, text = text, org = tuple(np.int32(approx[2][0] * .9)),
                  fontFace = font, fontScale = 1.2, color = (255,255,255), thickness = 2, lineType = cv2.LINE_AA)

plt.figure(figsize = (12, 6))
plt.imshow(cv2.cvtColor(res, cv2.COLOR_BGR2RGB))
plt.show()
```
